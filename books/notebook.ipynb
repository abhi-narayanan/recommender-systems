### Popularity Based Recommendation

```python
ratings_with_names = ratings.merge(books, on='ISBN')
```

```python
ratings_with_names.shape
```

```python
ratings_with_names.head()
```

```python
ratings_with_names.info()
```

```python
num_rating_df = ratings_with_names.groupby(by='Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
```

```python
num_rating_df.head()
```

```python
avg_rating_df = ratings_with_names.groupby(by=['Book-Title'])['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'}, inplace=True)
avg_rating_df.head()
```

```python
popularity_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popularity_df.head()
```

```python
popularity_df = popularity_df[popularity_df['num_ratings']>250]
```

```python
popularity_df = popularity_df.sort_values(by='avg_ratings', ascending=False).nlargest(n=50, columns='avg_ratings')
```

```python
popularity_df.merge(books, on='Book-Title').drop_duplicates(subset=['Book-Title']).shape
```

```python
books.columns
```

```python
popularity_df = popularity_df.merge(books, on='Book-Title').drop_duplicates(subset=['Book-Title'])[['Book-Title', 'Book-Author', 'Image-URL-M',
                                                                                   'num_ratings', 'avg_ratings']]
```

### Collaborative Filtering Based Recommendation

```python
ratings_with_names.head()
```

```python
# Users with minimum 200 reviews
x = ratings_with_names.groupby("User-ID").count()['Book-Rating'] > 200
active_users = x[x].index
```

```python
filtered_rating = ratings_with_names[ratings_with_names['User-ID'].isin(active_users)]
```

```python
filtered_rating.head()
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.head()
```

```python
pivot_table = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pivot_table.shape
```

```python
pivot_table.fillna(0, inplace=True)
```

```python
pivot_table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_score = cosine_similarity(pivot_table)
```

```python
print(sim_score)
```

```python
def recommend(book_name):
    # Index of book_name
    index = np.where(pivot_table.index == book_name)[0][0]
    distances = sim_score[index]
    similar_books = sorted(list(enumerate(distances)), key=lambda x: x[1], reverse=True)[1:6]

    data = []
    for i in similar_books:
        item = []
        temp_df = books[books['Book-Title'] == pivot_table.index[i[0]]]
        temp_df = temp_df.drop_duplicates('Book-Title')
        item.extend(list(temp_df['Book-Title'].values))
        item.extend(list(temp_df['Book-Author'].values))
        item.extend(list(temp_df['Image-URL-M'].values))

        data.append(item)

    return data
```
